**Average Votes By Years**

```python
imdb.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(data=imdb,x='Year',y='Votes',errorbar=None)
plt.title('Yearly Average Votes')
plt.show()
```

**Average Revenue By Year**

```python

imdb.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(data=imdb,x='Year',y='Revenue (Millions)',errorbar=None)
plt.title('Yearly Average Revenue')
```

**Director Average Ratings**

```python

imdb.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

```python
sns.barplot(data=imdb,x='Year',y='Rating',errorbar=None)
plt.title('Yearly Average Rating')
```

**Top 10 Movies By Runtime(Minutes)**

```python

imdbTop10=imdb[['Title','Runtime (Minutes)']].sort_values(by='Runtime (Minutes)',ascending=False).head(10)
imdbTop10
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=imdbTop10,x='Title',y='Runtime (Minutes)')
plt.title('Top 10 Movies By Runtime(Minutes)')
plt.xticks(rotation=90)
plt.show()
```

**Number of Movies Per Year**

**Top 10 Rated Movies**

```python
imdbTop10Rating=imdb[['Title','Director','Rating']].sort_values(by='Rating',ascending=False).head(10)
imdbTop10Rating
```

**Create Rating Category**

```python

def rating(rating):
  if rating>=7.0:
     return 'Excellent'
  elif rating>=6.0:
    return 'Good'
  else:
    return 'average'
imdb['Rating_Category']=imdb['Rating'].apply(rating)
imdb.head(5)
```
