```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().any()
```

```python
df.drop_duplicates()
```

**Nos .of Total  Male and Female Customers**

```python
cnt=df['Gender'].value_counts()
cnt
```

```python
px.bar(cnt,x=cnt.index,y=cnt.values,text_auto=True,labels={'index':'Gender','y':'counts'},
       title='Nos .of Total Male and Female Customers')
```

**Total Customers By Age Groups**

```python
age=df['Age Group'].value_counts()
age
```

```python
px.bar(age,x=age.index,y=age.values,text_auto=True,labels={'y':'Count of Age Groups','index':'Age Group'},title='Count of Age Groups')
```

**Total Male & Female Customers by Marital Status and Age Group**

```python
df['Status']=df['Marital_Status'].map({0:'Unmarried',1:'Married'})
df2=df.groupby (['Gender','Status','Age Group'],as_index=False)['Amount'].sum()
df2

px.sunburst(df2,path=['Gender','Status','Age Group'],values='Amount',color='Status',title='Total Male & Female Customers by Marital Status and Age Group')
```

**Total Sales By Gender and Age Group**

```python
from itertools import count
df1=df.groupby(['Gender','Age Group'],as_index=False)['Amount'].sum()
df1
```

```python
px.bar(df1,x='Age Group',y='Amount',color='Gender',text_auto='0.3s',height=600,title='Sales By Gender and Age Group')
```

**Top 10 States By No. of  Orders**

```python
state=df.groupby('State',as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
state
```

```python
px.bar(state,y='State',x='Orders',orientation='h',text_auto=True,title='Top 10 States By No. of Orders',)
```

**Total Sales By Occupation**

```python
occ=df.groupby('Occupation',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
occ
```

```python
px.bar(occ,x='Occupation',y='Amount',text_auto='0.3s',title='Total Sales By Occupation')
```

**Product Category by Amounts and Orders**

```python
cat=df.groupby('Product_Category',as_index=False).agg({'Orders':'count','Amount':'sum'})
cat
```

```python

px.bar(cat,x='Product_Category',y='Amount',color='Orders',log_y=True,barmode='group',text_auto='0.2s',title='Product_Category by Amounts and Orders')
```

**Total Amount and Orders by Customer**

```python
cust=df.groupby('Cust_name',as_index=False).agg({'Orders':'count','Amount':'sum'})
cust
```
